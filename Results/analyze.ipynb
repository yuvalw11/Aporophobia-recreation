```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = pd.read_csv(str('../my_results/agent_vars_42520_2:1_1.csv'), delimiter=";")
```

```python
x.query('status == "unemployed"')
```

This shows the profit along the simulation of 3 agents. one has a job, the second is a student, the third is unemployed.

```python
plt.plot(x.query('AgentID == 0')['wealth'])
plt.plot(x.query('AgentID == 97')['wealth'])
plt.plot(x.query('AgentID == 8')['wealth'])
```

This shows the profits of all agents during the 4 months. Some lost some gained

```python
profits = []
for i in range(100):
    profit = x.query(f'AgentID == {i}').iloc[-1]['wealth'] - x.query(f'AgentID == {i}').iloc[0]['wealth']
    profits.append(profit)

sns.histplot(data=profits, bins=20)
```
